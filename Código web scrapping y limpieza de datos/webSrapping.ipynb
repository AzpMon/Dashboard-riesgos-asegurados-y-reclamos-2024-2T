## WebScrapping Datos Población (segundo trimestre, Wikipedia)

```python
# Importar librerías para WebScraping y manipulación de datos
from bs4 import BeautifulSoup
import requests
import pandas as pd

# URL población México (wikipedia)
url = 'https://es.m.wikipedia.org/wiki/Anexo:Entidades_federativas_de_M%C3%A9xico_por_poblaci%C3%B3n'

# Realizar una solicitud GET para obtener el contenido HTML de la página
page = requests.get(url)

# Parsear el contenido HTML usando BeautifulSoup
soup = BeautifulSoup(page.text, 'html')

# Encontrar todas las tablas en el HTML
tablas = soup.find_all('table')

# Crear una lista para almacenar los DataFrames
dataframes = []

# Iterar sobre cada tabla encontrada
for i, tabla in enumerate(tablas):
    # Convertir la tabla HTML en un DataFrame de pandas
    df = pd.read_html(str(tabla))[0]
    
    # Añadir el DataFrame a la lista
    dataframes.append(df)
```

```python
# Se selecciona y muestra el datarframe de interés
poblacionSegundoTrimestre_df = dataframes[0]
poblacionSegundoTrimestre_df.head()
```

```python
# Se muestran las columnas del dataframe
poblacionSegundoTrimestre_df.columns
```

```python
# Se seleccionan los datos relevantes de la tabla
columnas_útiles=[
    ('Entidad federativa', 'Entidad federativa'), 
    ('Población', '2T2024'), 
    ('Contribución (%)', '2T2024')
    ]
poblacionSegundoTrimestre = poblacionSegundoTrimestre_df[columnas_útiles]

# Se renombran las columnas y se muestran
poblacionSegundoTrimestre.columns=['Estado', 'Población', 'Contribución']
poblacionSegundoTrimestre
```

```python
# Se guarda como un CSV 
poblacionSegundoTrimestre.to_csv('población.csv',index=False, encoding='utf-8')
```

## WebScrapping Datos PIB (Diciembre 2023,  Wikipedia)

```python
# URL del PIB por estados (wikipedia)
url2 = 'https://es.wikipedia.org/wiki/Anexo:Entidades_federativas_de_M%C3%A9xico_por_PIB'

# Realizar una solicitud GET para obtener el contenido HTML de la página
page2 = requests.get(url2)

# Parsear el contenido HTML usando BeautifulSoup
soup2 = BeautifulSoup(page2.text, 'html')

# Encontrar todas las tablas en el HTML
tablas2 = soup2.find_all('table')

# Crear una lista para almacenar los DataFrames
dataframes2 = []

# Iterar sobre cada tabla encontrada
for i, tabla in enumerate(tablas2):
    # Convertir la tabla HTML en un DataFrame de pandas
    df = pd.read_html(str(tabla))[0]
    
    # Añadir el DataFrame a la lista
    dataframes2.append(df)
```

```python
# Se selecciona y muestra el datarframe de interés
PIB_estados_df = dataframes2[0]
PIB_estados_df.head()
```

```python
PIB_estados_df.columns
```

```python
# Se seleccionan los datos relevantes de la tabla
columnas_útiles2=[
    ('Entidad federativa', 'Entidad federativa'), 
    ('PIB total', 'MXN'), 
    ('PIB total', '%')
    ]
PIB_estados = PIB_estados_df[columnas_útiles2]

# Se renombran las columnas y se muestran
PIB_estados.columns=['Estado', 'PIB_MXN', 'PIB_Porcentaje']
PIB_estados.head()
```

```python
# Se guarda el dataframe como un csv
PIB_estados.to_csv('PIB_estados.csv',index=False,  encoding='utf-8')
```
